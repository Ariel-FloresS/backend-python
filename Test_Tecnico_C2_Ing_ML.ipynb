2. ¿Cuáles consideras que son los tres mayores desafíos al pasar un modelo de un entorno de prueba a un entorno de producción?

Respuesta:

1. Reproducibilidad y consistencia:

    Asegurar que el modelo en producción se comporte igual que en pruebas, lo que implica versionar los  datos, features y modelo.

2. Escalabilidad:

    los pipelines de datos y entrenamiento del modelo deben soportar volúmenes reales de datos.

3. Monitoreo y mantenimiento:

    Implementar monitoreo para detectar drift y degradación del modelo, así como definir estrategias de reentrenamiento para mantener su desempeño en el tiempo.

3. ¿Qué es el "versionamiento" en MLOps y por qué es crucial? ¿Qué elementos específicos deberían ser versionados?

Respuesta: El versionamiento en MLOps es fundamental porque nos permite garantizar reproducibilidad, trazabilidad y auditoría a lo largo de todo el ciclo de vida del modelo. Es decir, poder responder preguntas como: ¿con qué datos se entrenó este modelo?, ¿qué configuración tenía?, ¿por qué este modelo funciona mejor que otro?

Para asegurar esta reproducibilidad, es importante versionar:

* Datos crudos (raw data): para poder reconstruir el punto de partida

* Datos limpios: ya que cualquier cambio en limpieza impacta el modelo

* Features generadas: especialmente si hay transformaciones complejas

* Datasets de entrenamiento: snapshot exacto usado para entrenar

* Modelo entrenado (artefacto): incluyendo su versión y configuración

* Métricas de evaluación: para comparar desempeño entre versiones


4. ¿Describe un escenario en el que podrías decidir implementar un modelo en nube o de forma local? ¿Qué factores influirían en esa decisión?

Respuesta:

* Modelo en Nube: Elegiría la nube cuando el caso de uso requiere procesar grandes volúmenes de datos o necesita escalabilidad. Por ejemplo, en un problema de forecasting con muchas series de tiempo, donde se requiere entrenamiento distribuido o procesamiento batch frecuente.

* Modelo Local: Optaría por una solución local cuando existen limitaciones de conectividad o latencia, como en plantas, agricultura o entornos industriales donde no se puede depender de internet.

5. ¿Cuál es la diferencia fundamental entre un pipeline de CI/CD para una aplicación de software tradicional y un pipeline de CI/CD para un modelo de Machine Learning (CI/CD for ML)?

Respuesta:

La diferencia principal es que en un pipeline de CI/CD tradicional solo se versiona, prueba y despliega código.

En cambio, en un pipeline de CI/CD para ML, además del código, también se deben versionar datos, features y modelos. Tambien, se incluyen procesos adicionales como validación de modelos (promoción basada en competencia) y la posibilidad de activar pipelines de reentrenamiento cuando se detecta drift.

6. El modelo está en producción y funcionando. ¿Qué tipos de monitoreo implementarías? Describe la diferencia entre monitorear métricas de sistema y métricas de rendimiento del modelo.

Respuesta:

Implementaría dos tipos de monitoreo:

1. Métricas de sistema:
Son métricas operativas del servicio, como latencia de inferencia, uso de CPU/GPU, memoria, throughput y errores. También incluiría monitoreo de logs del pipeline de datos para detectar fallas o anomalías (por ejemplo, datos faltantes o inconsistentes).

2. Métricas de rendimiento del modelo:
Evalúan la calidad del modelo:
* Con target: monitoreo de métricas como RMSE, MAPE, KS, etc.  para detectar degradación (concept drift).
* Sin target: monitoreo de data drift mediante PSI u otros métodos, así como cambios en la distribución o volumen de datos.

9. Describe cómo debería ser un proceso de re entrenamiento de un modelo de ML

Respuesta:
 
Una vez que identificamos deterioro en el modelo (por ejemplo, drift o caída en métricas), se ejecuta un pipeline de reentrenamiento con la siguiente lógica:

Preparación y versionamiento de datos:

1. Se toman los datos crudos más recientes y se procesan mediante los pipelines de limpieza y feature engineering:

     Estos datos se versionan para asegurar trazabilidad y reproducibilidad.

2. Split de datos (train / validation / test):

Se genera un nuevo dataset dividido en:

* train: para ajustar el modelo
* Test : para ajustar hiperparámetros y evitar overfitting
* validation: para evaluar el desempeño final de forma imparcial

3. Entrenamiento del modelo (challenger):

Se entrena un nuevo modelo utilizando el conjunto de entrenamiento.

4. Evaluación y competencia (champion vs challenger):

El modelo challenger se compara contra el modelo en producción (champion) bajo las siguientes condiciones:

* No presenta overfitting/underfitting (consistencia entre train y tets)
* Mejora o mantiene el desempeño en el conjunto de test (RMSE, MAPE, etc.)
* Tiene un comportamiento estable entre validation y test

5. Reentrenamiento final:

Si el challenger es mejor, se rentrenará utilizando todos los datos disponibles (train + validation + test) 

6. Promoción a producción:

El modelo se versiona y se despliega como nuevo modelo productivo.



10. Has implementado un modelo que, sin darte cuenta, introduce un sesgo (bias) que afecta negativamente a un grupo demográfico de clientes. ¿Cómo podrías, desde la ingeniería y el MLOps, haber mitigado el riesgo de que esto ocurriera y qué harías para remediarlo una vez detectado?

Respuesta:

Para mitigar el riesgo de bias, implementaría validaciones en varias etapas:

* Antes del entrenamiento: analizar el dataset o usar herramientas como IBM Fairness 360 (framework opensource de IBM para dtectar bias en el datatset) para evaluar métricas de fairness por grupo (si tenemos conjuntos de datos desvalanceados, etc).

* Antes de producción: usar técnicas de explainability (SHAP Values, Feature Importance) para entender que features son importantes para el modelo y poder identificar si las features utilizadas por el modelo  no tienen sesgo (por ejemplo que el modelo le de peso a el genero de una persona).

* En produccion: seguiria monitoreando metricas de Fairness de los nuevos datos (datos input de nuestrop modelo en producciion)

Si detecto bias en producción:

* Corregiría el dataset o las features
* Reentrenaría el modelo con controles de fairness 
* Como medida temporal, podría hacer rollback a una versión anterior, pero solo si se valida que esa versión no presenta el mismo problema
